## 3. Argumentos Nomeados (Keyword Arguments)

Conceitos abordados: chamada de funções passando argumentos pelo nome do parâmetro

Argumentos nomeados permitem passar valores em qualquer ordem, tornando as chamadas mais legíveis e menos suscetíveis a erros de posição. Podem ser combinados com argumentos posicionais, desde que os posicionais venham primeiro.

### Exercício 3.1

1. Passando todos os argumentos por nome em ordem invertida

3. Usando o valor padrão de `copia_oculta`

````
def enviar_email(destinatario, assunto, corpo, copia_oculta=None):
    print(f'Para: {destinatario}')
    print(f'Assunto: {assunto}')
    print(f'Corpo: {corpo}')
    if copia_oculta:
        print(f'CCO: {copia_oculta}')
    print('---')

# Escreva as três chamadas abaixo
````

Um sistema de consulta de banco de dados possui a função abaixo. Observe que ela tem 5 parâmetros, sendo 3 opcionais. Faça chamadas que demonstrem por que argumentos nomeados tornam o código mais legível quando há muitos parâmetros opcionais — experimente passar apenas `tabela` e `limite`, depois `tabela`, `condicao` e `ordem`.

````
def consultar_bd(tabela, condicao=None, campos='*', ordem=None, limite=100):
    query = f'SELECT {campos} FROM {tabela}'
    if condicao:
        query += f' WHERE {condicao}'
    if ordem:
        query += f' ORDER BY {ordem}'
    query += f' LIMIT {limite}'
    print(query)

# Escreva as chamadas abaixo
````
