## 🔀 Bloco 1 — `if`, `elif`, `else`

**Exercício 1** — Verificar velocidade

```python
velocidade = 135

if velocidade > 120:
    print("Você está acima da velocidade! 🚨")
else:
    print("Velocidade dentro do limite. ✅")
```

**Exercício 2** — Saudação por hora do dia

```python
hora = 14

if hora <= 11:
    print("Bom dia! 🌅")
elif hora <= 17:
    print("Boa tarde! ☀️")
else:
    print("Boa noite! 🌙")
```

**Exercício 3** — Comparar dois valores

```python
a = 42
b = 27

if a > b:
    print(f"O maior é {a}")
elif a == b:
    print("Os valores são iguais")
else:
    print(f"O maior é {b}")
```

**Exercício 4** — Login com `and`

```python
usuario = "admin"
senha = 1234

if usuario == "admin" and senha == 1234:
    print("Acesso liberado! 🔓")
else:
    print("Acesso negado! 🔒")
```

**Exercício 5** — Busca em lista com `in`

```python
estoque = ["caneta", "caderno", "borracha", "régua", "lápis"]
produto_buscado = "caderno"

if produto_buscado in estoque:
    print(f"Produto encontrado! ✅")
else:
    print(f"Produto não encontrado. ❌")
```

## 🔁 Bloco 2 — `for`

**Exercício 6** — Saudar cada nome da lista

```python
colegas = ["Ana", "Bruno", "Carol", "Diego", "Elisa"]

for nome in colegas:
    print(f"Olá, {nome}!")
```

**Exercício 7** — Números de 1 a 20 com `range()`

```python
for i in range(1, 21):
    print(i, end=" ")
```

**Exercício 8** — Múltiplos de 3 entre 1 e 30

```python
for i in range(1, 31):
    if i % 3 == 0:
        print(i, end=" ")
```

**Exercício 9** — Somar lista de preços

```python
precos = [12.50, 8.00, 35.90, 4.75, 22.00]
total = 0

for preco in precos:
    total += preco

print(f"Total: R$ {total:.2f}")
```

**Exercício 10** — Exibir palavras em maiúsculas

```python
palavras = ["python", "lista", "laço", "código", "aprender"]

for palavra in palavras:
    print(palavra.upper())
```

**Exercício 11** — `enumerate()` com posição e nome

```python
ingredientes = ["feijão", "arroz", "macarrão", "sal", "azeite"]

for posicao, ingrediente in enumerate(ingredientes):
    print(f"{posicao + 1}: {ingrediente}")
```

## 🔄 Bloco 3 — `while`

**Exercício 12** — Contar de 1 a 10 com `while`

```python
contador = 1

while contador <= 10:
    print(contador, end=" ")
    contador += 1
```

**Exercício 13** — Contagem regressiva

```python
regressiva = 10

while regressiva >= 1:
    print(regressiva, "...")
    regressiva -= 1

print("🎉 Feliz Ano Novo!")
```

**Exercício 14** — Soma dos pares de 2 a 20

```python
numero = 2
soma = 0

while numero <= 20:
    soma += numero
    numero += 2

print("Soma dos pares de 2 a 20:", soma)
```

**Exercício 15** — Esvaziar fila com `while`

```python
fila = ["Alice", "Bruno", "Carol", "Diego"]

while len(fila) > 0:
    pessoa = fila.pop(0)
    print(f"Atendendo: {pessoa}")

print("\nFila vazia!", fila)
```

## 🌟 Bloco 4 — Desafios Mistos

**Exercício 16** ⭐ — Separar pares e ímpares

```python
numeros = [3, 8, 15, 4, 7, 12, 19, 6, 11, 20]
pares   = []
impares = []

for num in numeros:
    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)

print("Pares  :", pares)
print("Ímpares:", impares)
```

**Exercício 17** ⭐ — Classificar notas

```python
notas = [8.0, 4.5, 6.0, 9.5, 5.0, 3.0, 7.0]

for nota in notas:
    if nota >= 7:
        status = "Aprovado ✅"
    elif nota >= 5:
        status = "Recuperação ⚠️"
    else:
        status = "Reprovado ❌"
    print(f"Nota {nota:.1f} → {status}")
```

**Exercício 18** ⭐ — Tabuada do 8

```python
for i in range(1, 11):
    print(f"8 x {i:2} = {8 * i}")
```

**Exercício 19** ⭐ — Maior e menor sem `max()` / `min()`

```python
valores = [34, 12, 78, 5, 56, 90, 23, 47]
maior = valores[0]
menor = valores[0]

for v in valores:
    if v > maior:
        maior = v
    if v < menor:
        menor = v

print("Lista :", valores)
print("Maior :", maior)
print("Menor :", menor)
```

**Exercício 20** ⭐⭐ — Analisar lista de dicionários

```python
produtos = [
    {"nome": "Caderno",     "preco": 25.90},
    {"nome": "Mochila",     "preco": 129.00},
    {"nome": "Caneta",      "preco": 3.50},
    {"nome": "Calculadora", "preco": 79.90},
    {"nome": "Borracha",    "preco": 2.00},
    {"nome": "Estojo",      "preco": 55.00},
]

quantidade_caros = 0
total = 0
caros = []

for produto in produtos:
    total += produto["preco"]
    if produto["preco"] > 50:
        quantidade_caros += 1
        caros.append(produto["nome"])

print(f"a) Produtos acima de R$ 50,00: {quantidade_caros}")
print(f"b) Valor total: R$ {total:.2f}")
print(f"c) Produtos caros: {caros}")
```
