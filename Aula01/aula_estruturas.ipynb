## 📋 1. Listas
> Coleções **ordenadas** e **mutáveis** — usamos colchetes `[ ]`

```python
# Criando listas
frutas  = ["maçã", "banana", "laranja", "uva", "manga"]
numeros = [10, 20, 30, 40, 50]
mista   = [1, "Python", 3.14, True]   # pode misturar tipos

print("Frutas :", frutas)
print("Números:", numeros)
print("Mista  :", mista)
```

```python
# Acessando elementos pelo índice (começa em 0!)
print("Primeiro elemento :", frutas[0])   # maçã
print("Último elemento   :", frutas[-1])  # manga  (índice negativo)
print("Segundo elemento  :", frutas[1])   # banana

print("\nQuantidade de frutas:", len(frutas))
```

## ⚙️ 2. Operações com Listas

```python
cores = ["vermelho", "azul", "verde"]
print("Original:", cores)

cores.append("amarelo")          # adiciona ao FINAL
print("append('amarelo'):", cores)

cores.insert(1, "roxo")          # insere na posição 1
print("insert(1,'roxo') :", cores)

cores.remove("azul")             # remove pelo VALOR
print("remove('azul')   :", cores)

retirado = cores.pop(0)          # remove pelo ÍNDICE
print(f"pop(0) → retirou '{retirado}':", cores)
```

```python
# Ordenação
bagunca = [5, 2, 8, 1, 9, 3]
bagunca.sort()
print("Crescente :", bagunca)

bagunca.sort(reverse=True)
print("Decrescente:", bagunca)
```

```python
# count · in · + · *
letras = ["a", "b", "a", "c", "a"]
print("'a' aparece", letras.count("a"), "vezes")

frutas = ["maçã", "banana", "laranja", "uva", "manga"]
print("'banana' em frutas?", "banana" in frutas)
print("'melão'  em frutas?", "melão"  in frutas)

lista1 = [1, 2, 3]
lista2 = [4, 5, 6]
print("Concatenação:", lista1 + lista2)
print("Repetição   :", ["epa"] * 3)
```

## ✂️ 3. Slicing (Fatiamento)
> Sintaxe: `lista[início : fim : passo]`  
> O índice **fim** **não** é incluído!

```python
abc = ["a", "b", "c", "d", "e", "f", "g"]
print("Lista completa:", abc)
print("Índices       :", list(range(len(abc))))

print()
print("abc[0:3]  →", abc[0:3])   # índices 0, 1, 2
print("abc[2:5]  →", abc[2:5])   # índices 2, 3, 4
print("abc[:4]   →", abc[:4])    # do início até 3
print("abc[3:]   →", abc[3:])    # do índice 3 até o fim
print("abc[:]    →", abc[:])     # cópia completa
```

```python
print("Com passo:")
print("abc[::2]  →", abc[::2])   # de 2 em 2
print("abc[1::2] →", abc[1::2])  # começa em 1, de 2 em 2
print("abc[::-1] →", abc[::-1])  # invertida!

print("\nÍndices negativos:")
print("abc[-3:]  →", abc[-3:])   # últimos 3
print("abc[:-2]  →", abc[:-2])   # tudo exceto os 2 últimos
```

## 🔒 4. Tuplas
> Coleções **ordenadas** e **imutáveis** — usamos parênteses `( )`

```python
coordenadas = (10, 20)
cores_rgb   = (255, 128, 0)
notas       = (7, 8, 9, 7, 10, 7)

print("Coordenadas:", coordenadas)
print("RGB        :", cores_rgb)
print("Notas      :", notas)

print("\nx =", coordenadas[0], "| y =", coordenadas[1])
```

```python
# Tupla unitária precisa de vírgula!
unitaria  = (42,)
nao_tupla = (42)
print("Com vírgula :", unitaria,  "→", type(unitaria))
print("Sem vírgula :", nao_tupla, "→", type(nao_tupla))
```

```python
notas = (7, 8, 9, 7, 10, 7)
print("Nota 7 aparece", notas.count(7), "vezes")
print("Índice do 10  :", notas.index(10))

# Desempacotamento
nome, idade, cidade = ("Ana", 20, "São Paulo")
print(f"\nNome: {nome} | Idade: {idade} | Cidade: {cidade}")
```

```python
# Tuplas são IMUTÁVEIS — isso causa erro:
try:
    coordenadas = (10, 20)
    coordenadas[0] = 99
except TypeError as e:
    print("TypeError:", e)
```

## 📖 5. Dicionários
> Pares **chave: valor** — mutáveis, ordenados (Python ≥ 3.7) — usamos `{ }`

```python
aluno = {
    "nome"  : "Carlos",
    "idade" : 21,
    "curso" : "Ciência da Computação",
    "notas" : [8.5, 9.0, 7.5]
}

print("Dicionário:", aluno)
print("\nNome  :", aluno["nome"])
print("Curso :", aluno["curso"])
print("Notas :", aluno["notas"])
```

```python
# get → sem erro se a chave não existir
print("Matrícula:", aluno.get("matricula", "não informada"))

# Adicionar e modificar
aluno["email"] = "carlos@email.com"
aluno["idade"] = 22
print("\nAtualizado:", aluno)

# Remover
del aluno["email"]
print("Sem email :", aluno)
```

```python
print("Chaves  :", list(aluno.keys()))
print("Valores :", list(aluno.values()))
print("Itens   :", list(aluno.items()))

print("\n'nome'  no dicionário?", "nome"  in aluno)
print("'email' no dicionário?", "email" in aluno)
```

## 🔵 6. Sets (Conjuntos)
> **Não ordenados**, **sem duplicatas** — usamos `{ }` (sem chave:valor)

```python
vogais = {"a", "e", "i", "o", "u"}
print("Vogais:", vogais)

# Duplicatas são removidas automaticamente!
com_rep = {1, 2, 2, 3, 3, 3, 4}
print("Com repetição {1,2,2,3,3,3,4} →", com_rep)
```

```python
vogais.add("y")
print("add('y')    :", vogais)

vogais.discard("y")        # discard não gera erro se não existir
print("discard('y'):", vogais)
```

```python
turma_a = {"Ana", "Bruno", "Carlos", "Diana"}
turma_b = {"Carlos", "Diana", "Eduardo", "Flávia"}

print("Turma A:", turma_a)
print("Turma B:", turma_b)

print("\nUnião (A | B)       :", turma_a | turma_b)
print("Interseção (A & B)  :", turma_a & turma_b)
print("Diferença (A - B)   :", turma_a - turma_b)
print("Dif. simétrica (^)  :", turma_a ^ turma_b)
```
